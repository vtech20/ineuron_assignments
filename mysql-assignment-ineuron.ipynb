```python
#2 Create a database called as cardataset
cur.execute("create database cardataset")
connection.commit()
cur.execute("Show databases")
cur.fetchall()
```

```python
#3 create a table called as car with a column name given in data set descriptions
cur.execute('create table cardataset.car(buying VARCHAR(20) ,maint VARCHAR(20)  ,doors VARCHAR(20) ,persons VARCHAR(20) ,lug_boot VARCHAR(20), safety VARCHAR(20), classes VARCHAR(20))') 
connection.commit()
cur.execute("use cardataset")
cur.execute("Show tables")
cur.fetchall()
```

```python
#4 dump all the data into car table

import csv
with open('car.data','r') as f:
    car_data=csv.reader(f,delimiter='\n')
    for i in car_data:
        print(f'insert into cardataset.car values ({str(i[0])})')
        cur.execute('insert into cardataset.car(buying,maint,doors,persons,lug_boot,safety,classes) values(%s,%s,%s,%s,%s,%s,%s)',i[0].split(","))
connection.commit()
```

```python
#5 try to check whether all the data is available inside your table or not
cur.execute('select * from cardataset.car')
cur.fetchall()
```

```python
#6 try to group all the dat with col1 and count occurances of each and every record based on col1 value

cur.execute('select buying,count(buying) from cardataset.car group by buying')
cur.fetchall()
```

```python
#7 try to filter a record where col3 value will be 4
cur.execute('select * from cardataset.car where doors = 4')
cur.fetchall()
```

```python
# 8. Try to update a col 3 value with 8 wherever you have value equal to 2

cur.execute('update cardataset.car set doors = "8" where doors = "2"')
connection.commit()
cur.execute('select * from cardataset.car')
cur.fetchall()
```

```python
# 9. Try to delete table
cur.execute('drop table cardataset.car')
connection.commit()
cur.execute('use cardataset')
cur.execute('show tables')
cur.fetchall()
```

```python
# 10. try to delete database
cur.execute('drop database cardataset')
connection.commit()
cur.execute('show databases')
cur.fetchall()
```
